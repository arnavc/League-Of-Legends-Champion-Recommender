```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist, jaccard, cosine
from sklearn.metrics import jaccard_similarity_score

path = '/Users/arnavc/galvanize/capstone/data/util_freq_matrix.csv'
df_main = pd.read_csv(path)
#df_main.index.astype(int)
```

```python
df_main.index = df_main.index.map(int)
```

```python
user_id =df_main['SummonerId']
df = df_main.drop('SummonerId',axis=1,inplace=False)
df_main=df_main[:500]
df=df[:500]
```

```python
df.head()
df.shape
type(df.index)
```

```python
data_ibs = pd.DataFrame(index=df.columns, columns=df.columns)
data_ibs.shape
```

```python
for i in range(0, len(data_ibs.columns)):
    for j in range(0, len(data_ibs.columns)):
        data_ibs.iloc[i,j] = 1 - cosine(df.iloc[:,i],df.iloc[:,j])
```

```python
neighbors = pd.DataFrame(index=data_ibs.columns, columns = range(1,11))
```

```python
data_ibs.head()
```

```python
for i in range(0,len(data_ibs.columns)):
    neighbors.iloc[i,:10] = (data_ibs.iloc[0:,i]).sort_values(ascending=False)[:10].index
```

```python
neighbors.head()
```

```python
import csv
def champid_to_name():
    path = '/Users/arnavc/galvanize/capstone/data/league2016/ChampId2Name.csv'
    reader = csv.reader(open(path))

    result = {}
    for row in reader:
        key = row[0]
        result[key] = row[1:]

    return result
```

```python
champ_dict = champid_to_name()
```

```python
neighbors_name = neighbors.copy()
print(neighbors_name.head())
print(neighbors.head())
```

```python
for col in neighbors_name.columns:
    neighbors_name[col].replace(champ_dict, inplace=True)
```

```python
neighbors_name.head(100)
```

```python
def get_score(history, similarities):
    return (sum(history * similarities) / sum(similarities))
```

```python
data_sims = pd.DataFrame(index=df_main.index,columns=df_main.columns)
data_sims.iloc[:,:1] = df_main.iloc[:,:1]
```

```python
len(data_sims.index)
```

```python
for i in range(0,len(data_sims.index)):
    for j in range(1,len(data_sims.columns)):
        user = data_sims.index[i]
        product = data_sims.columns[j]
 
        if df_main.iloc[i][j] == 1:
            data_sims.iloc[i][j] = 0
        else:
            product_top_names = neighbors.loc[product][1:10]
            product_top_sims = data_ibs.loc[product].sort_values(ascending=False)[1:10]
            user_purchases = df.loc[user,product_top_names]
 
            data_sims.iloc[i][j] = get_score(user_purchases,product_top_sims)
```

```python
data_recommend = pd.DataFrame(index=data_sims.index, columns=['user','1','2','3','4','5','6','7','8'])
data_recommend.iloc[0:,0] = data_sims.iloc[:,0]
```

```python
for i in range(0,len(data_sims.index)):
    data_recommend.iloc[i,1:] = data_sims.iloc[i,:].sort_values(ascending=False).iloc[1:9,].index.transpose()
```
